### Bar Plot
digunakan untuk :
1. membandingkan nilai antar kategori

```python
kategori = ['Python', 'assembly', 'java', 'C++', 'Java Script']
Jumlah_pengguna = [100, 10, 50, 70, 90]

plt.figure(figsize=(8,5))
# plt.bar
plt.bar(kategori, Jumlah_pengguna,color="Blue", label="Perbandingan bahasa dan user")

# tambah elemen
plt.title("Jumlah Pengguna")
plt.xlabel("Bahasa Pemrograman")
plt.ylabel("Jumlah Pengguna")
plt.grid(axis='y', alpha = 0.5, linestyle = "--")
plt.legend()
plt.show()
```

```python
kategori = ['Batch 1', 'Batch 2', 'Batch 3']
Laki_laki = [40, 60, 55]
perempuan = [15, 30, 50]



x = np.arange(len(kategori)) # [0,1,2,]
barwidth = 0.35



plt.figure(figsize=(8,5))
# plt.bar
# plt.bar(kategori, Jumlah_pengguna,color="Blue", label="Perbandingan bahasa dan user")
plt.bar(x - barwidth/2, Laki_laki, width = barwidth, label = " Laki Laki")
plt.bar(x + barwidth/2, perempuan, width = barwidth, label="Perempuan")
        


# tambah elemen
plt.title("Jumlah Pengguna")
plt.xlabel("Bahasa Pemrograman")
plt.ylabel("Jumlah Pengguna")
plt.grid(axis='y', alpha = 0.5, linestyle = "--")
plt.legend()
plt.xticks(x, kategori)

# menyesuaikan secara otomatis layoutnya biar konsisten
plt.tight_layout()
plt.show()
```

### Scatter plot
Digunakan untuk :
1. Menampilkan hubungan antara 2 variabel atau lebih

```python
umur = [18, 21, 25, 30, 35, 50, 45, 50]
penghasilan = [1.5, 2.2, 2.8, 4.0, 4.2, 4.5, 5.0, 3.5]

plt.figure(figsize=(8,5))
plt.scatter(umur, penghasilan)

plt.title("Hubungan antara umur dan penghasilan")
plt.xlabel("Umur (Tahun)")
plt.ylabel('Penghasilan (Juta)')
plt.grid(True)

plt.tight_layout()
plt.show()

```

```python
fisika_A = [75, 80, 85, 90, 40]
matematika_A = [70, 85, 82, 90, 100]

fisika_B = [60, 65, 20, 60, 78]
matematika_B = [40, 30, 40, 89, 10]

plt.scatter(fisika_A, matematika_A, marker = "o", label="Nilai kelas A")
plt.scatter(fisika_B, matematika_B, marker = "*", label="Nilai Kelas B")

plt.title("Hubungan antara Fisika dan Matematika")
plt.xlabel("Fisika")
plt.ylabel('Matematika')
plt.grid(True, alpha = 0.5)

plt.legend(loc="lower left")
plt.tight_layout()
plt.show()

```

### Box Plot

```python

plt.boxplot(df["Total Amount"],vert=False)
plt.xlabel('Age Group')
plt.ylabel('Number of Customer')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
views = [320, 450, 400, 390, 410, 470, 500, 490, 480, 510,
         530, 600, 620, 590, 550, 570, 610, 640, 700, 5000]

plt.boxplot(views,vert=False)
plt.title("Outlier pada views")
plt.xlabel('Number pada views')
plt.ylabel('Instagram')
plt.grid(axis='x', alpha=0.3)
plt.show()
```

```python
views_Series = pd.Series(views)
views_clean = views_Series[views_Series < 1000]
```

```python
plt.Figure(figsize=(10,6))
plt.boxplot(views_clean,vert=False)
plt.title("Outlier pada views")
plt.xlabel('Number pada views')
plt.ylabel('Instagram')
plt.grid(axis='x', alpha=0.3)
plt.show()
```

### Subplot

```python
hari = ['Sen', 'Sel', 'Rab', 'Kam', 'Jum', 'Sab', 'Min']
followers = [120, 130, 145, 160, 175, 190, 210]
views = [300, 400, 350, 500, 600, 550, 700]
likes = [100, 120, 110, 160, 200, 180, 250]
comments = [20, 25, 22, 35, 40, 38, 50]
engagement_rate = [0.4, 0.36, 0.38, 0.39, 0.4, 0.39, 0.43]

# plt.subplot
fig, axs = plt.subplots(2,2,  figsize = (12, 8))

axs[0, 0].plot(hari, followers, marker = "o")
axs[0, 0].set_title("Pertumbuhan Followers")
axs[0, 0].set_ylabel("Jumlah Followers")
axs[0, 0].grid(alpha=0.3)

axs[0, 1].bar(hari, views, color = "skyblue")
axs[0, 1].set_title("Pertumbuhan views")
axs[0, 1].set_ylabel("Jumlah Followers")
axs[0, 1].grid(alpha=0.3)

axs[1, 0].plot(hari, likes, color="orange", marker = "o")
axs[1, 0].set_title("Pertumbuhan Likes")
axs[1, 0].set_ylabel("Jumlah likes")
axs[1, 0].grid(alpha=0.3)

axs[1, 1].scatter(hari, engagement_rate, color="purple", marker = "o")
axs[1, 1].set_title("Pertumbuhan engagement_rate")
axs[1, 1].set_ylabel("Jumlah engagement_rate")
axs[1, 1].grid(alpha=0.3)
axs[1, 1].set_ylim(0,0.5)

fig.suptitle("Trafic Media Sosial", fontsize = "24")
plt.tight_layout()
```

### Correlation Matrix

```python
df.head()
```

```python
df.columns
```

```python
numerical_df = df[['Age','Quantity','Price per Unit','Total Amount']]
```

```python
import seaborn as sns
```

```python
correlation_matrix = numerical_df.corr()
correlation_matrix
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(
    correlation_matrix,
    fmt=".2f",
    cmap='coolwarm',
    annot=True,
    linewidths=-.5,
    cbar_kws={'label': r"Koefisien Korelasi"}
)
plt.title("Matrix Correlation", fontsize=20)
plt.show()
```
