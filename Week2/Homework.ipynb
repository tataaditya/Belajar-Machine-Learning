#### Aggregation dan Feature Engineering

```python
# Pastikan kolom yang akan dianalisis sudah dibersihkan
# Misalnya: df['site_review_rating'], df['guest_recommendation'], df['site_review_count'] sudah diisi NaN-nya

# Agregasi hotel per provinsi berdasarkan rating dan jumlah review
agg_df = DataFrame.groupby(['province', 'property_name']).agg({
    'site_review_rating': 'mean',
    'guest_recommendation': 'mean',
    'site_review_count': 'sum'
}).reset_index()

# Buat kolom skor gabungan (opsional)
agg_df['score'] = (
    agg_df['site_review_rating'].fillna(0) * 0.6 +
    agg_df['guest_recommendation'].fillna(0) * 0.4
)

# Urutkan dari yang terbaik
agg_df_sorted = agg_df.sort_values(by=['province', 'score', 'site_review_count'], ascending=[True, False, False])

# Tampilkan hotel terbaik tiap provinsi
top_hotels_by_province = agg_df_sorted.groupby('province').head(1)
print(top_hotels_by_province)
```

#### Bersihkan Outlier

```python
numerical_cols = DataFrame.select_dtypes(include=['int64','float64']).columns.to_list()
```

```python
def remove_outliers_iqr(df,columns,X):
    df_clean = df.copy()
    for col in columns :
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - X * IQR
        upper_bound = Q3 + X * IQR
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    return df_clean
```

```python
df_no_outlier = remove_outliers_iqr(DataFrame, numerical_cols,1)
```

```python
df_no_outlier
```

```python
# Minmax Scaling, mengubah data menjadi rentang dari 0 - 1
from sklearn.preprocessing import MinMaxScaler

df_minmax_scaled = df_no_outlier.copy()
```

```python
min_max_scale = MinMaxScaler()
```

```python
df_minmax_scaled[numerical_cols] = min_max_scale.fit_transform(df_no_outlier[numerical_cols])
```
