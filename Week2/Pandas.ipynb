```python
numerical_cols = DataFrameNew.select_dtypes(include=['int64','float64']).columns.to_list()
```

```python
def remove_outliers_iqr(df,columns,X):
    df_clean = df.copy()
    for col in columns :
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - X * IQR
        upper_bound = Q3 + X * IQR
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    return df_clean
```

```python
df_no_outlier = remove_outliers_iqr(DataFrameNew, numerical_cols,1)
```

```python
df_no_outlier
```

```python
# Minmax Scaling, mengubah data menjadi rentang dari 0 - 1
from sklearn.preprocessing import MinMaxScaler

df_minmax_scaled = df_no_outlier.copy()
```

```python
min_max_scale = MinMaxScaler()
```

```python
df_minmax_scaled[numerical_cols] = min_max_scale.fit_transform(df_no_outlier[numerical_cols])
```

```python
df_minmax_scaled.describe()
```

```python
# Standard Scaling, untuk membuat mean menjadi 0 dan standar deviation menjadi 1
from sklearn.preprocessing import StandardScaler

stf_scaler = StandardScaler()
```

```python
std_scaler = StandardScaler()
```

```python
df_std_scaled = df_no_outlier.copy()
df_std_scaled[numerical_cols] = std_scaler.fit_transform(df_no_outlier[numerical_cols])
```
